## Importing necessary libraries


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Data preprocessing


```python
credit_card_data = pd.read_csv('/content/creditcardfraud.zip')
credit_card_data.head()
```

```python
credit_card_data.sample()
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data['Class'].value_counts()
```

**Assigning class name to data (fraud-1)and(legit-0)**

```python
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data['Class']==1]
```

```python
fraud['Class']
```

```python
legit.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

**creating new dataset with only two class name**

```python
new_df = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

## Splitting the preprocessed data in training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

## Using logistic regression


```python
model=LogisticRegression()
```

```python
model.fit(X_train, Y_train)
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy on Training data : ', training_data_accuracy)
```

## Model evaluation

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score on Test Data : ', test_data_accuracy)
```
